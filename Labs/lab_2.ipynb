Цикли

```python
# 1. Цикл for з range
letters = ["a", "b", "c"]
print("--- Цикл FOR з range ---")
for i in range(len(letters)):
    print(f"На позиції {i} знаходиться буква {letters[i]}")
else:
    print("Цикл завершено успішно (блок else виконано).")

# 2. Цикл WHILE (підрахунок до 5)
print("\n--- Цикл WHILE ---")
count = 0
while count < 5:
    print(f"Поточний лічильник: {count}")
    count += 1
```

5. Розгалуження

```python
from random import randint

# 1. Розгалуження if/elif/else
temperature = randint(-5, 10)

print(f"Поточна температура: {temperature}°C")
if temperature < 0:
    print("Дуже холодно!")
elif temperature >= 0 and temperature < 5:
    print("Прохолодно, одягніть куртку.")
else:
    print("Тепло, можна гуляти!")

# 2. Тернарний оператор (з прикладу завдання)
A = randint(0, 1)
print("-" * 30)
print(f"Значить А={A}" if A else "Але може бути що А={}".format(A))
```

6. Конструкція try -> except -> finally

```python
A = 0
B = 10
try:
    print(f"Спроба ділення: {B}/{A}")
    # Тут виникне ZeroDivisionError
    result = B / A
    print(f"Результат: {result}")
except ZeroDivisionError as e:
    # Обробляємо лише конкретний тип помилки
    print("Невже це помилка > ZeroDivisionError:", e)
except Exception as e:
    # Обробляємо будь-яку іншу помилку
    print("Інша невідома помилка > ", e)
finally:
    print(">> Блок 'finally' завжди виконується, незалежно від помилок!")
```

Контекст-менеджер with

```python
# Припустимо, у файлі README.md є кілька рядків тексту
print("--- Читання файлу за допомогою 'with' ---")
try:
    with open("README.md", "r") as f:
        print(f"Файл відкритий? {f.closed == False}")
        for index, line in enumerate(f):
            if index < 3: # Читаємо лише перші 3 рядки
                print(f"Рядок {index+1}: {line.strip()}")
            else:
                break
except FileNotFoundError:
    print("Помилка: Файл README.md не знайдено.")

# Після виходу з блоку with, ресурс закривається автоматично
# Спроба доступу до f.closed після блоку with:
# print(f"Файл закритий? {f.closed}")
```

Python Lambdas

```python
# 1. Створення лямбди
# Лямбда: приймає first та age, повертає відформатований рядок.
this_is_lambda = lambda first, age: f'Створив: {first}, Мені {age} років'

# 2. Виклик лямбди
print("Це її виклик:", this_is_lambda('Олег', 20))

# 3. Використання лямбди як ключа сортування
students = [('Олег', 18), ('Анна', 22), ('Іван', 20)]
# Сортуємо список за віком (елемент з індексом 1)
sorted_students = sorted(students, key=lambda student: student[1])

print(f"\nСтуденти за віком: {sorted_students}")
```
