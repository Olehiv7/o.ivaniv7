```python

class MyName:
    """Опис класу / Документація
    """
    total_names = 0 #Class Variable

    def __init__(self, name=None) -> None:
        """Ініціалізація класу
        """
        self.name = name if name is not None else self.anonymous_user().name #Class attributes / Instance variables
        MyName.total_names += 1 #modify class variable
        self.my_id = self.total_names

    @property
    def whoami(self) -> str: 
        """Class property
        return: повертаємо імя 
        """
        return f"My name is {self.name}"
    
    @property
    def my_email(self) -> str:
        """Class property
        return: повертаємо емейл
        """
        return self.create_email()
    
    def create_email(self) -> str:
        """Instance method
        """
        return f"{self.name}@itcollege.lviv.ua"

    @classmethod
    def anonymous_user(cls):
        """Classs method
        """
        return cls("Anonymous")
    
    @staticmethod
    def say_hello(message="Hello to everyone!") -> str:
        """Static method
        """
        return f"You say: {message}"


print("Розпочинаємо створювати обєкти!")

names = ("Marta", None, "Oleh")
all_names = {name: MyName(name) for name in names}

for name, me in all_names.items():
    print(f"""{">*<"*20}
This is object: {me} 
This is object attribute: {me.name} / {me.my_id}
This is {type(MyName.whoami)}: {me.whoami} / {me.my_email}
This is {type(me.create_email)} call: {me.create_email()}
This is static {type(MyName.say_hello)} with defaults: {me.say_hello()} 
This is class variable {type(MyName.total_names)}: from class {MyName.total_names} / from object {me.total_names}
{"<*>"*20}""")

print(f"We are done. We create {me.total_names} names! ??? Why {MyName.total_names}?")
```

Чому коли передаємо значення None створюється обєкт з іменем Anonymous?

Якщо значення, передане аргументу name, є None, виконується частина else self.anonymous_user().name. Метод anonymous_user (який є класовим методом @classmethod) створює новий екземпляр класу, передаючи йому рядок "Anonymous".

Допишіть функцію в класі яка порахує кількість букв імені (підказка: використайте функцію len());

```python

class MyName:
    """Опис класу / Документація
    """
    total_names = 0 #Class Variable

    def __init__(self, name=None) -> None:
        """Ініціалізація класу
        """
        self.name = name if name is not None else self.anonymous_user().name #Class attributes / Instance variables
        MyName.total_names += 1 #modify class variable
        self.my_id = self.total_names

    @property
    def whoami(self) -> str: 
        """Class property
        return: повертаємо імя 
        """
        return f"My name is {self.name}"
    
    @property
    def my_email(self) -> str:
        """Class property
        return: повертаємо емейл
        """
        return self.create_email()
    
    def create_email(self) -> str:
        """Instance method
        """
        return f"{self.name}@itcollege.lviv.ua"

    @classmethod
    def anonymous_user(cls):
        """Classs method
        """
        return cls("Anonymous")
    
    @staticmethod
    def say_hello(message="Hello to everyone!") -> str:
        """Static method
        """
        return f"You say: {message}"
    @property
    def name_length(self) -> int:
      """Повертає кількість символів в імені."""
      return len(self.name)

print("Розпочинаємо створювати обєкти!")

names = ("Bohdan", "Marta", None, "Oleh")
all_names = {name: MyName(name) for name in names}

for name, me in all_names.items():
    print(f"""{">*<"*20}
This is object: {me} 
This is object attribute: {me.name} / {me.my_id}
This is {type(MyName.whoami)}: {me.whoami} / {me.my_email}
This is {type(me.create_email)} call: {me.create_email()}
This is static {type(MyName.say_hello)} with defaults: {me.say_hello()} 
This is class variable {type(MyName.total_names)}: from class {MyName.total_names} / from object {me.total_names}
{"<*>"*20}""")
marta = all_names.get("Marta") # Отримуємо об'єкт
if marta:
    print(f"Змінене привітання: {marta.say_hello('Вітаю вас від імені Марти!')}")
print(f"We are done. We create {me.total_names} names! ??? Why {MyName.total_names}?")
print(f"Довжина імені Marta: {all_names['Marta'].name_length}")
```

Додайте перевірку: якщо ім'я містить цифри або символи, підніміть помилку ValueError("Ім'я може містити лише літери!").

```python
# В КЛАС MYNAME, МЕТОД __init__
def __init__(self, name=None) -> None:
    # ...
    raw_name = name if name is not None else self.anonymous_user().name
    
    # <-- МОДИФІКАЦІЯ: ДОДАЄМО ПЕРЕВІРКУ
    if not raw_name.isalpha():
        raise ValueError("Ім'я може містити лише літери!")

    self.name = raw_name.capitalize()
    MyName.total_names += 1
    self.my_id = self.total_names
```

Додайте нову властивість full_name, яка повертає результат у форматі: "User #<id>: <name> (<email>)"

```python
# ВСТАВИТИ В КЛАС MYNAME
@property
def full_name(self) -> str:
    """Нова властивість, що повертає повну інформацію про користувача."""
    return f"User #{self.my_id}: {self.name} ({self.my_email})"
```

Реалізуйте метод save_to_file(filename="users.txt"), який додає рядок із записом у файл.

```python
# ВСТАВИТИ В КЛАС MYNAME
def save_to_file(self, filename="users.txt"):
    """Записує повну інформацію про користувача у вказаний файл."""
    record = self.full_name + "\n"
    with open(filename, 'a', encoding='utf-8') as f:
        f.write(record)
    print(f"Запис '{self.name}' додано до файлу {filename}")
```
